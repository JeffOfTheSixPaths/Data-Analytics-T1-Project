```python
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3 as sql
```

We make a connection to the database and then list out all of the shows we're going to look at


```python
con = sql.connect("WestminsterDogShow.db")

shows = ['Herding', 'Hound', 'InShow', 'NonSporting', 'Sporting', 'Terrier', 'Toy', 'Working']
```

Later in the notebook, I'm going to get all of the dog groups that have won a competition.

In order to do this, I can make a dictionary where the key is the breed and the output from the dictionary is the group of the breed inputed

I can join together the breed and group tables and make a dictionary. Then set each breed as the key and the group as the value.

```python
query = 'select * from breeds as b inner join groups as g on b.groupId = g.groupId'
breeds_and_groups = pd.read_sql_query(query ,con)

breeds_to_groups = {}
for i in range(breeds_and_groups.shape[0]):
    breeds_to_groups[breeds_and_groups['name'][i]] = breeds_and_groups['groupName'][i]
```

the query function gets a database that joins the winners of a show and all the characteristics of the dog, which is stored in V_doge

```python
def query(show: str):
    return f'select * from best{show} as s inner join V_doge as d on s.dogId = d.dogId'

```

The commentary on the graphs will be after the graphs are shown

```python
herding = pd.read_sql_query(query(shows[0]), con)
count_breed = herding[['name:1']].value_counts()


count_group = herding['groupName'].value_counts()


count_breed.plot.bar()
plt.show()
count_group.plot.bar()
```
